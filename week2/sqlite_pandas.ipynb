```python
import sqlite3
import pandas as pd
```

## Setup SQlite connection

```python
cnx = sqlite3.connect(':memory:') #load database in memory
```

```python
df_daily = pd.read_csv(
    "https://raw.githubusercontent.com/thoo/covid-tracking-data/master/data/states_daily_4pm_et.csv"
)
```

```python
df_daily.head()
```

### Load data into sql database 

```python
df_daily.to_sql(name='covid_daily',con=cnx)
```

### Query with pandas 

```python
pd.read_sql('select * from covid_daily limit 10',con=cnx)
```

### Query directly from the connection

```python
cur = cnx.cursor()
```

```python
cur.execute('select * from covid_daily limit 10')
```

```python
rows=cur.fetchall()
```

```python
for row in rows:
    print(row)
```

```python
pd.DataFrame(rows)
```
